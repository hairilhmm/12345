```python
import pandas as pd
import googletrans
from googletrans import Translator
```

```python
df = pd.read_csv('Green Paradise Pagar Alam - Salin - Salin.csv')
df.head(10)
```

```python
df = pd.read_excel('Kawasan_Wisata_Gunung_Dempo_1.xlsx')
df.head(10)
```

```python
translator = Translator()
translations = {}
for column in df.columns:
  unique_elements = df[column].unique()
  for element in unique_elements:
    translations[element] = translator.translate(element).text
translations
```

```python
df.replace(translations, inplace = True)
df.head(10)
```

```python
df.replace(translations, inplace = True)
df.head(10)
```

```python
df.to_csv('translite_Kawasan_Wisata_Gunung_Dempo_1.csv',encoding='utf8', index=False)
```
